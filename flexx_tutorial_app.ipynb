# Tutorial for flexx.app - connecting to the browser

```python
from flexx import app, event
```

In normal operation, one uses `app.launch()` to fire up a browser (or desktop app) to run the JavaScript in. This is followed by `app.run()` (or `app.start()` for servers) to enter Flexx' main loop.

In the notebook, however, there already is a browser (and a mainloop). To tell Flexx that we're in the notebook, use `app.init_notebook()` at the start of your notebook.

```python
app.init_notebook()
```

```python
class MyModel(app.Model):
    
    class Both:
        @event.prop
        def foo(self, v=''):
            return str(v)
    
    class JS:
        
        @event.connect('foo')
        def on_foo(self, *events):
            if self.foo:
                window.alert('foo is ' + self.foo)
```

```python
m = MyModel()
```

```python
m.foo = 'hello'
```
